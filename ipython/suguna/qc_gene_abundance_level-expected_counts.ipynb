### Deliverable #1: Genes that are mapped in every single nuclei based on expected count

```python
rsem_dir = os.path.join(data_dir, 'HCN_RSEM_Tophat-Cufflinks_analysis_data','RSEM')
single_nuclei_fns = [os.path.join(rsem_dir,filename)
                    for filename in os.listdir(rsem_dir)
                    if filename.split('-')[1].split('_')[0] == '1' and filename.find('genes') != -1
                    and not filename.startswith('A1-')]
```

```python
def read_nuclei(filename):
    sample_name = os.path.basename(filename).split('.')[0]
    df = pd.read_table(filename, index_col=0)
    df['expected_count_'+sample_name] = df['expected_count']
    return df[['expected_count_'+sample_name]]
```

```python
combined_df = pd.concat([read_nuclei(fn)
                         for fn in single_nuclei_fns],
                        axis=1)
```

```python
tpm_columns = [column for column in combined_df.columns
               if column.find('expected_count') != -1]
```

```python
def expressed_in_all(series):
    for column in tpm_columns:
        if series[column] == 0:
            return False
    return True
```

```python
combined_df['mean'] = combined_df[tpm_columns].mean(axis=1)
combined_df['std'] = combined_df[tpm_columns].std(axis=1)
```

```python
expressed = combined_df.apply(expressed_in_all, axis=1)
```

```python
expressed_df = combined_df[expressed]
```

```python
expressed_df
```

### Only one gene expressed in all. Let's instead just tell the number of samples, it is expressed in.

```python
def expressed_in_x(series, tpm_columns):
    num_expressed = 0
    for column in tpm_columns:
        if series[column] != 0:
            num_expressed += 1
    return num_expressed
```

```python
combined_df['Expressed in x'] = combined_df.apply(expressed_in_x,
                                                  args=[tpm_columns],
                                                  axis=1)
```

```python
combined_df.sort('Expressed in x', ascending=False).head()[['mean','std','Expressed in x']]
```

```python
writer = pd.ExcelWriter(os.path.join(data_dir, 'Expression in Single Nuclei - Expected Counts.xlsx'))
column_order = ['Expressed in x','mean','std']+tpm_columns
combined_df.sort('Expressed in x', ascending=False)[column_order].to_excel(writer)
```

```python
writer.close()
```

### Deliverable #2: Bulk RNA analysis

```python
bulk_nuclei_fns = [os.path.join(rsem_dir,filename)
                    for filename in os.listdir(rsem_dir)
                    if filename.split('-')[1].split('_')[0] != '1' and filename.find('genes') != -1]
combined_bulk_df = pd.concat([read_nuclei(fn)
                             for fn in bulk_nuclei_fns],
                             axis=1)
tpm_bulk_columns = [column for column in combined_bulk_df.columns
                    if column.find('TPM') != -1]
combined_bulk_df['mean'] = combined_bulk_df[tpm_bulk_columns].mean(axis=1)
combined_bulk_df['std'] = combined_bulk_df[tpm_bulk_columns].std(axis=1)
combined_bulk_df['Expressed in x'] = combined_bulk_df.apply(expressed_in_x,
                                                            args=[tpm_bulk_columns],
                                                            axis=1)
```

```python
combined_bulk_df.sort(['Expressed in x', 'mean'], ascending=False).head()[['mean','std','Expressed in x']]
```

```python
writer = pd.ExcelWriter(os.path.join(data_dir, 'Expression in Bulk Nuclei.xlsx'))
column_order = ['Expressed in x','mean','std']+tpm_bulk_columns
combined_bulk_df.sort(['Expressed in x','mean'], ascending=False)[column_order].to_excel(writer)
```

```python
writer.close()
```

### Find interesting genes by top expressed, illumina bodymap, and housekeeping genes

```python
import mygene
mg = mygene.MyGeneInfo()
```

```python
def get_common_name(ensemble_id):
    ensemble_id = ensemble_id.split('.')[0]
    try:
        return mg.getgene(ensemble_id, 'name,symbol,refseq')['symbol']
    except AssertionError:
        return ensemble_id
    except:
        try:
            return mg.getgene(ensemble_id, 'name,symbol,refseq')[0]['symbol']
        except:
            return ensemble_id
    
        
```

```python
combined_df['gene_name'] = combined_df.index.map(get_common_name)
```
