### Load index statistics

```python
mapped_stats = {}
for filename in os.listdir(main_dir):
    if filename.endswith('.idxstats'):
        sample = filename.split('.')[0]
        filename = os.path.join(main_dir, filename)
        df = pd.read_table(filename, sep='\t',
                           names=['contig', 'length', 'reads_mapped', 'reads_unmapped'])
        ercc_mapped = df[df.contig.str.contains('ERCC')]['reads_mapped'].sum()
        other_mapped = df[~df.contig.str.contains('ERCC')]['reads_mapped'].sum()
        unmapped = df['reads_unmapped'].sum()
        mapped_stats[sample] = (ercc_mapped, other_mapped, unmapped)
```

```python
mapped_df = pd.DataFrame(mapped_stats).T
mapped_df.columns = ['ERCC', 'GENOME', 'UNMAPPED']
```

### These commands were run on SDSC

```python
read_count_dir = '../../data/suguna/read_counts/gtf_based/'
read_count_files = dict((filename.split('_')[0], os.path.join(read_count_dir, filename))
                         for filename in os.listdir(read_count_dir))
```

```python
read_count_files
```

```python
def load_read_counts(filename):
    if filename.find('intergenic') != -1:
        bed_columns = ['chromosome', 'start', 'end']
    elif filename.find('exonic') != -1:
        bed_columns = ['chromosome', 'start',' end', 'count', 'identifiers']
    elif filename.find('intronic') != -1:
        bed_columns = ['chromosome', 'start',' end', 'identifiers']
        
    columns = bed_columns + sample_names
    df = pd.read_table(filename, names=columns)
    
    # Remove ERCC
    if filename.find('intergenic') == -1:
        df = df[~df.identifiers.str.contains('ERCC')]
    
    read_df = pd.DataFrame(df[df.columns[len(bed_columns):]].sum())
    return df, read_df
```

```python
read_dfs = dict((genomic_type, load_read_counts(filename))
                for genomic_type, filename in read_count_files.iteritems())
```

```python
read_df = pd.concat([value[1] for value in read_dfs.values()], axis=1)
```

```python
read_df.columns = ['Intergenic', 'Intronic', 'Exonic']
```

```python
single_nuclei = read_df.index.map(lambda x: x.split('-')[1][0] == '1')
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(211)
ax = read_df[single_nuclei].plot(kind='barh', stacked=True, ax=ax)
ax.set_title('Single Nuclei')
ax = fig.add_subplot(212)
ax = read_df[~single_nuclei].plot(kind='barh', stacked=True, ax=ax)
ax.set_title('Bulk Nuclei')
fig.tight_layout()
```
