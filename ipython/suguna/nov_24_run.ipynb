```python
out_template = '../../pipeline/nov_24_run.yaml'
run_info = {}
run_info['experiment_date'] = 'Nov 24 2014'
run_info['experiment_name'] = 'Various Sample Info'
run_info['output_dir'] = tscc_data+'/nov_24_run/'
samples_dir = tscc_data+'/nov_24_run/fastq'
sample_info = {}
```

```python
with open(out_template, 'w') as OUT:
    with open(run_template_yaml) as RUN:
        for line in RUN:
            for key, value in run_info.items():
                line = line.replace(key, value)
            OUT.write(line)
    for i in range(0, len(sample_files), 2):
        sample_info['file1'] = os.path.join(samples_dir, sample_files[i])
        sample_info['file2'] = os.path.join(samples_dir, sample_files[i+1])
        sample_info['sample_name'] = sample_files[i].split('_')[1]
        with open(sample_template_yaml) as SAMPLE:
            for line in SAMPLE:
                for key, value in sample_info.items():
                    line = line.replace(key, value)
                OUT.write(line)
        
```
