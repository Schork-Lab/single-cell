### Numbers of Genes Found vs Number of Mapped Reads

```python
pratap_qc = "HBPC_RNASEQ_QC_PV_12012014.xlsx"
pratap_df = pd.read_excel(pratap_qc, sheetname="RSEM", skiprows=1, index_col=0).ix[:23]
```

```python
samtools_qc = "ercc_comparison.xlsx"
samtools_df = pd.read_excel(samtools_qc)
```

```python
merged_df = samtools_df.merge(pratap_df, left_index=True, right_index=True)
```
