```python
def read_sample_tpms(filename):
    sample_name = os.path.basename(filename).split('.')[0]
    df = pd.read_table(filename, index_col=0)
    df[sample_name] = df['TPM']
    return df[[sample_name]]
```

```python
sample_dir = os.path.join(local_data, 'nov_24_run', 'HBPC_RSEM_RNASeq_PV_12012014',
                          'expression_data', 'genes')
```

```python
sample_dfs = {}
for sample_fn in os.listdir(sample_dir):
    if sample_fn.endswith('.results'):
        sample_name = sample_fn.split('.')[0]
        sample_dfs[sample_name] = read_sample_tpms(os.path.join(sample_dir, sample_fn))
```

```python
sample_tpms = pd.concat(sample_dfs.values(), axis=1)
```

```python
control = '3DHCN100pg111314'
```

```python

control_tpm = sample_tpms[control]
control_expressed = control_tpm[control_tpm > 0]
print "Total expressed in control: ", len(control_expressed)
```

```python
print "Binning into three by expressions"
low_expressed = control_expressed[control_expressed <= control_expressed.quantile(0.33)]
mid_expressed = control_expressed[(control_expressed > control_expressed.quantile(0.33)) &
                                  (control_expressed < control_expressed.quantile(0.66)) ]
high_expressed = control_expressed[(control_expressed >= control_expressed.quantile(0.66)) ]
print "Low Expressed", control_expressed.quantile(0.33), len(low_expressed)
print "Mid Expressed", control_expressed.quantile(0.66), len(mid_expressed)
print "High Expressed", control_expressed.max(), len(high_expressed)
```

```python
sample_number_expressed = {}
for sample in sample_tpms.columns:
    sample_tpm = sample_tpms[sample]
    sample_expressed = sample_tpm[sample_tpm > 0]
    sample_expressed_genes = set(sample_expressed.index)
    low_intersection = len(sample_expressed_genes.intersection(low_expressed.index)) 
    mid_intersection = len(sample_expressed_genes.intersection(mid_expressed.index)) 
    high_intersection = len(sample_expressed_genes.intersection(high_expressed.index)) 
    novel_expressed = len(sample_expressed_genes.difference(control_expressed.index))
    sample_number_expressed[sample] = (low_intersection, mid_intersection,
                                       high_intersection, novel_expressed)
    
```

```python
expressed_df = pd.DataFrame(sample_number_expressed).T
expressed_df.columns = ['Low', 'Mid', 'High', 'Novel']
```

```python
expressed_df.plot(kind='bar')
```

```python
percentage_df = expressed_df[['Low','Mid','High']].astype(float)/expressed_df[['Low','Mid','High']].ix[control]
```

```python
def type_of_rna_prep(x):
    num_of_cells = x.split('HCN')[0][2:] 
    if num_of_cells == '':
        return x.split('HCN')[1].split('pg')[0]+'pg'
    return num_of_cells
```

```python
percentage_df['RNA Source'] = percentage_df.index.map(type_of_rna_prep)
percentage_df['Date'] = percentage_df.index.map(lambda x: x[-6:])
```

```python
expressed_df['RNA Source'] = expressed_df.index.map(type_of_rna_prep)
expressed_df['Date'] = expressed_df.index.map(lambda x: x[-6:])
```

### Create low expressed gtf

```python
gtf_file = os.path.join(local_data, 'reference', 'GRCh37_ERCC_GFP_RNASpikes.gtf')
```

```python
low_expressed_genes = set(low_expressed.index)
low_expressed_out = os.path.join(local_data, 'reference', 'GRCh37_ERCC_GFP_RNASpikes.low_expressed.gtf')
with open(gtf_file) as IN:
    with open(low_expressed_out, 'w') as OUT:
        for line in IN:
            try:
                gene = "ENSG"+line.split("ENSG")[1].split('"')[0]
                if gene in low_expressed_genes:
                    OUT.write(line)
            except:
                continue
            
```

### Create high expressed gtf / mid expressed

```python
high_expressed_genes = set(high_expressed.index)
high_expressed_out = os.path.join(local_data, 'reference', 'GRCh37_ERCC_GFP_RNASpikes.high_expressed.gtf')
mid_expressed_genes = set(mid_expressed.index)
mid_expressed_out = os.path.join(local_data, 'reference', 'GRCh37_ERCC_GFP_RNASpikes.mid_expressed.gtf')
with open(gtf_file) as IN:
    with open(high_expressed_out, 'w') as HIGH_OUT:
        with open(mid_expressed_out, 'w') as MID_OUT:
            for line in IN:
                try:
                    gene = "ENSG"+line.split("ENSG")[1].split('"')[0]
                    if gene in high_expressed_genes:
                        HIGH_OUT.write(line)
                    elif gene in mid_expressed_genes:
                        MID_OUT.write(line)
                except:
                    continue
            
```
