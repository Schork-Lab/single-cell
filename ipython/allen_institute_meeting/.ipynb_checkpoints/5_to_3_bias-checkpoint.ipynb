```python
all_samples = {'HC Single Cell 1': '2B1HCNPN111314',
               'HC Single Cell 2': '2C1HCNPN111314',
               'LC Single Cell 1': '2A1HCNPN111314',
               'LC Single Cell 2': '1E1HCNP111314',
               'Control': '3DHCN100pg111314'}
good_samples = {'HC Single Cell 1': '2B1HCNPN111314',
               'HC Single Cell 2': '2C1HCNPN111314',
               'LC Single Cell 1': '2A1HCNPN111314',
               #'LC Single Cell 2': '1E1HCNP111314',
               'Control': '3DHCN100pg111314'}
```

```python
qc_dir = os.path.join(local_data, 'nov_24_run', 'qc', 'pratap')
fn = os.path.join(qc_dir, 'rseqc-combined.geneBodyCoverage.txt')
```

```python
def load_coverage(fn):
    df = pd.read_table(fn, index_col=0).T
    df.columns = df.columns.map(lambda x: x.split('.')[0][1:])
    df_range = df.max()-df.min()
    normalized_df = (df - df.min())/df_range
    
    return normalized_df, df

def create_plot(df, samples, gene_name, coverage_type='Relative Coverage'):
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111)
    sample_names, sample_ids = zip(*sorted(samples.items()))
    ax = df[list(sample_ids)].plot(ax=ax, lw=3)
    legend = ax.legend(sample_names, loc=0)
    for label in legend.get_texts():
        label.set_size(12)
    labels = ax.set_xticklabels(['','20','40','60','80'])
    ax.set_xlabel("Gene Body (5\' -> 3\')", fontdict={'size':14})
    ax.set_ylabel(coverage_type, fontdict={'size':14})
    ax.set_title('%s' % gene_name, fontdict={'size':16})
    return fig

def create_heatmap(normalized_df, gene_name):
    heatmap_df = normalized_df.T
    def type_of_rna_prep(x):
        num_of_cells = x.split('HCN')[0][2:] 
        if num_of_cells == '':
            return x.split('HCN')[1].split('pg')[0]+'pg'
        return num_of_cells
    heatmap_df['RNA Source'] = heatmap_df.index.map(type_of_rna_prep)
    heatmap_df['Date'] = heatmap_df.index.map(lambda x: x[-6:])
    data_df = heatmap_df.sort(['Date','RNA Source']).T.ix[:100].T
    fig = plt.figure(figsize=(12,9))
    ax = fig.add_subplot(111)
    sns.set(context="notebook")
    sns.heatmap(data_df.astype(float), linewidths=0, cmap=plt.get_cmap('RdBu'), ax=ax)
    ax.set_xticks(range(0, 100, 20))
    ax.set_xticklabels(['0', '20', '40', '60', '80'], rotation=0)
    ax.set_xlabel('Transcript Body 5\'-> 3\'', fontdict={'size':14})
    ax.set_title('%s' % gene_name, fontdict={'size':16})
    return fig
```

```python
figure_dir = os.path.join(qc_dir, 'figures')
relative_dir = os.path.join(figure_dir, 'relative_coverage')
absolute_dir = os.path.join(figure_dir, 'absolute_coverage')
heatmaps_dir = os.path.join(figure_dir, 'heatmaps')

if not os.path.exists(figure_dir): os.mkdir(figure_dir)
if not os.path.exists(relative_dir): os.mkdir(relative_dir)
if not os.path.exists(absolute_dir): os.mkdir(absolute_dir)
if not os.path.exists(heatmaps_dir): os.mkdir(heatmaps_dir)
for fn in os.listdir(qc_dir):
    if fn.endswith('geneBodyCoverage.txt'):
        fn = os.path.join(qc_dir, fn)
        gene_name = fn.split('-')[1].split('.')[0].replace('_', ' ').upper()
        normalized_df, coverage_df = load_coverage(fn)
        
        print gene_name
        # Using Absolute Coverage
        fig = create_plot(coverage_df, good_samples, gene_name, 'Absolute Coverage');
        fig.savefig(os.path.join(absolute_dir, '%s-good-samples.png' % gene_name))
        fig = create_plot(coverage_df, all_samples, gene_name, 'Absolute Coverage');
        fig.savefig(os.path.join(absolute_dir, '%s-all-samples.png' % gene_name))
        
        # Using Relative Coverage
        fig = create_plot(normalized_df, good_samples, gene_name);
        fig.savefig(os.path.join(relative_dir, '%s-good-samples.png' % gene_name))
        fig = create_plot(normalized_df, all_samples, gene_name);
        fig.savefig(os.path.join(relative_dir, '%s-all-samples.png' % gene_name))
        
        # Heatmaps
        fig = create_heatmap(normalized_df, gene_name)
        fig.savefig(os.path.join(heatmaps_dir, '%s-all-samples.png' % gene_name))
```
