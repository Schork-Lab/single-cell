### Finally works, so now let's try the coverageBed tool again

```python
cmd = "bamToBed -splitD -i %s | awk \'BEGIN{OFS=\"\\t\"}$1=\"chr\"$1\' | bedtools coverage -a - -b %s \
    > %s" % (sample_bam_top, intron_file, sample_bam+'.intronic_coverage.tophat')
```

```python
print local_to_tscc(cmd)
```

```python
intron_coverage = sample_bam+'.intronic_coverage.tophat'
columns = ['Chromosome', 'Start', 'End', 'GTF-INFO', 'Reads', 'Basesgt0', 'Length', 'Fractiongt0']
intron_df = pd.read_table(intron_coverage, names=columns)
```

```python
intron_covered = intron_df[(intron_df.Reads > 0) & (intron_df.Length > 100)]
```

```python
%pylab inline
import seaborn as sns
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(intron_covered['Fractiongt0'], intron_covered['Length'])
ax.set_ylim([0, 10000])
ax.set_xlim([0, 1])
ax.set_ylabel('Length of Intron', fontdict={'size': 14})
ax.set_xlabel('Fraction of Intron Covered with >0 Reads', fontdict={'size': 14})
ax.set_title('Coverage Consistency Across Introns for 2B1HCNPN111314', fontdict={'size':16})
fig.savefig('Coverage_Consistency_Fraction_Covered_2B1HCNPN111314.png')
```
