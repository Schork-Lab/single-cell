```python
import pandas as pd
```

```python
import seaborn as sns
%pylab inline
```

```python
def get_mapped_df(idx_dir):
    mapped_stats = {}
    for filename in os.listdir(idx_dir):
        if filename.endswith('.idxstats'):
            sample = filename.split('.')[0]
            filename = os.path.join(idx_dir, filename)
            df = pd.read_table(filename, sep='\t',
                               names=['contig', 'length', 'reads_mapped', 'reads_unmapped'])
            ercc_mapped = df[df.contig.str.contains('ERCC')]['reads_mapped'].sum()
            other_mapped = df[df.contig.str.contains('ERCC')]['reads_mapped'].sum()
            unmapped = df['reads_unmapped'].sum()
            mapped_stats[sample] = (ercc_mapped, other_mapped, unmapped)
    mapped_df = pd.DataFrame(mapped_stats).T
    mapped_df.columns = ['ERCC', 'GENOME', 'UNMAPPED']
    return mapped_df
```

```python
mapped_df = get_mapped_df(nov24_idx_dir)
```

```python
def type_of_rna_prep(x):
    num_of_cells = x.split('HCN')[0][2:] 
    if num_of_cells == '':
        return x.split('HCN')[1].split('pg')[0]+'pg'
    return num_of_cells
```

```python
mapped_df['RNA Source'] = mapped_df.index.map(type_of_rna_prep)
mapped_df['Date'] = mapped_df.index.map(lambda x: x[-6:])
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(211)
ax = mapped_df.sort(['Date', 'RNA Source'], ascending=False).plot(kind='barh', stacked=True, ax=ax)
ax.set_title('All Samples - November 24', fontdict={'size': 14})
ax.set_xlim([0, max(mapped_df.sum(axis=1))]);
fig.tight_layout()
```

```python
initial_idx_dir = os.path.join(local_data, 'initial_run', 'rsem_bams')
initial_mapped_df = get_mapped_df(initial_idx_dir)
```

```python
def type_of_rna_prep_initial(x):
    num_of_cells = x.split('_')[0].split('-')[1]
    if num_of_cells == 'RNA':
        return x.split('_')[0].split('-')[2]
    return num_of_cells
```

```python
initial_mapped_df['Date'] = '08012014'
initial_mapped_df['RNA Source'] = initial_mapped_df.index.map(type_of_rna_prep_initial)
```

```python
ax = fig.add_subplot(212)
ax = initial_mapped_df.sort('RNA Source', ascending=False).plot(kind='barh', stacked=True, ax=ax)
ax.set_title('All Samples - Initial Run', fontdict={'size': 14})
ax.set_xlim([0, max(mapped_df.sum(axis=1))]);
ax.set_xlabel('Number of Reads')
fig.tight_layout()
```
