### This is a little weird as this is a sort of normalized transcript count. In the methods they go deeper into converting the UMI to the number of transcripts. There are 256 possible UMIs that they tag each read with and then they use this UMI number to go back to the # of transcript count based on the math below

For each gene $i$, $k_{o,i}$ denotes the number of observed UMIs and $k_{n,i}$ the number of non-observed UMIs. As such, the total number K of UMIs is given by 

\begin{equation}
K = k_{o,i} + k_{n,i}
\end{equation}

The probability of not observing $k_{n,i}$ UMIs for a gene with $m_i$ copies is given by 
\begin{equation}
(1-1/K)^{m_i} = \dfrac{k_{n,i}}{K}
\end{equation}

If we assume each UMI has an equal probability of being mapped to a read, that is each UMI: k has a 1/K chance of being tagged to a read. The probability of a particular k not being tagged in m reads is (1-1/K)^m . Therefore for $k_{n,i}$ UMIs to not map, the probability should be 
\begin{equation}
(1-\dfrac{1}{K})^{m_i * k_{n,i}}
\end{equation} and not what they postulate
