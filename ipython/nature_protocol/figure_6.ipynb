```python
def load_coverage(fn):
    df = pd.read_table(fn, index_col=0).T
    df.columns = df.columns.map(lambda x: x.split('.')[0][1:])
    df_range = df.max()-df.min()
    normalized_df = (df - df.min())/df_range   
    return normalized_df, df
```

```python
fn = os.path.join(qc_dir, 'rseqc-high_expressed.geneBodyCoverage.txt')
```

```python
normalized_df, coverage_df = load_coverage(fn)
```
