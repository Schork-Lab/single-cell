### Use DESeq Results

```python
deseq_values = pd.read_csv('../../../data/deseq_adjusted_high_expression.csv',
                           index_col=0)
deseq_values = deseq_values.T
```

```python
deseq_values
```

```python
pca = PCA(n_components=len(deseq_values))
pca.fit(deseq_values)
projections = pca.fit_transform(deseq_values)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(projections[:6,0], projections[:6,1], c=get_sorted_color_map()[:6], alpha=0.4, label='Neuronal')
ax.scatter(projections[6:10,0], projections[6:10,1], c=get_sorted_color_map()[6:10], alpha=0.4, label='Non-neuronal')
ax.scatter(projections[10:14,0], projections[10:14,1], c=get_sorted_color_map()[10:14], alpha=0.4, label='Pooled')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.legend()
```

```python
projections_df = pd.DataFrame(projections)
projections_df['Type'] = ['Neuronal']*6+['Non-neuronal']*4+['Pooled']*4
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(projections[:6,0], projections[:6,1], projections[:6,2], 
           c=get_sorted_color_map()[:6], alpha=0.8, label='Neuronal')
ax.scatter(projections[6:10,0], projections[6:10,1], projections[6:10,2], 
           c=get_sorted_color_map()[6:10], alpha=0.8, label='Non-neuronal')
ax.scatter(projections[10:14,0], projections[10:14,1], projections[10:14,2],
           c=get_sorted_color_map()[10:14], alpha=0.8, label='Pooled')
ax.set_xlabel('PC 1 (%0.2f)' % (pca.explained_variance_ratio_[0]*100))
ax.set_ylabel('PC 2 (%0.2f)' % (pca.explained_variance_ratio_[1]*100))
ax.set_zlabel('PC 3 (%0.2f)' % (pca.explained_variance_ratio_[2]*100))
ax.legend(loc=0)
```

### Use TPM Values

```python
low, mid, high = data.get_low_mid_high_genes(tpm_df['Total RNA-100pg-1'], 
                                             cutoffs=(0.33, 0.66))
high_tpm_df = tpm_df.ix[high.index]
high_tpm_df = high_tpm_df.T
pca = PCA(n_components=len(high_tpm_df))
pca.fit(high_tpm_df)
projections = pca.fit_transform(high_tpm_df)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(projections[:6,0], projections[:6,1], c=get_sorted_color_map()[:6], alpha=0.4, label='Neuronal')
ax.scatter(projections[6:10,0], projections[6:10,1], c=get_sorted_color_map()[6:10], alpha=0.4, label='Non-neuronal')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.legend()
```
