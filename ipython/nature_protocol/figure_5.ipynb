```python
gs = gridspec.GridSpec(1, 2)
```

```python
mapped_df = data.load_mapped_data()
mapped_df = data.paperify_df(mapped_df.T).T
```

```python
fig = plt.figure(figsize=(8,4), dpi=300)
ax = fig.add_subplot(gs[0])
ax = mapped_df.ix[get_sorted_order()].plot(kind='barh', stacked=True, ax=ax)
sns.despine()
ax.grid(False, axis='y')
#ax.set_title('Mapping Read Statistics')
ax.set_xlabel('Number of Reads')
```

```python
tpm_df = data.filter_df(data.load_tpms(), expressed_in_multiple=True, genes_only=True)
neuronal_tpm = data.filter_df(tpm_df, non_neuronal=False, pooled=False)
non_neuronal_tpm = data.filter_df(tpm_df, neuronal=False, pooled=False)
pooled_tpm = data.filter_df(tpm_df, neuronal=False, non_neuronal=False)
expressed_value_counts = {}
expressed_value_counts['Neuronal'] = sum(neuronal_tpm > 0, axis=1).value_counts()[1:]
expressed_value_counts['Non-neuronal'] = sum(non_neuronal_tpm > 0, axis=1).value_counts()[1:]
expressed_value_counts['Pooled'] = sum(pooled_tpm > 0, axis=1).value_counts()[1:]
```

```python
from matplotlib_venn import venn3
```

```python
set1 = set(neuronal_tpm[sum(neuronal_tpm > 0, axis=1) > 0].index)
set2 = set(non_neuronal_tpm[sum(non_neuronal_tpm > 0, axis=1) > 0].index)
set3 = set(pooled_tpm[sum(pooled_tpm > 0, axis=1) > 0].index)
ax = fig.add_subplot(gs[1])
venn3([set1, set2, set3], ('Neuronal', 'Non-neuronal', 'Pooled'), ax=ax)
plt.show()
```
