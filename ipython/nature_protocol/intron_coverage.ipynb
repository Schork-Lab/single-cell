### Let's compare this to the exons coverage.

```python
exon_file = os.path.join(local_data, "reference", "gtf_based",
                           "GRCh37_ERCC_GFP_RNASpikes.gtf.exons.bed")
```

```python
cmd = "bamToBed -splitD -i %s | awk \'BEGIN{OFS=\"\\t\"}$1=\"chr\"$1\' | bedtools coverage -a - -b %s \
    > %s" % (sample_bam_top, exon_file, sample_bam+'.exon_coverage.tophat')
```

```python
print local_to_tscc(cmd)
```

```python
exon_coverage = sample_bam+'.exon_coverage.tophat'
columns = ['Chromosome', 'Start', 'End', 'GTF-INFO', 'Reads', 'Basesgt0', 'Length', 'Fractiongt0']
exon_df = pd.read_table(exon_coverage, names=columns)
```

```python
exon_covered = exon_df[(exon_df.Reads > 0) & (exon_df.Length > 100)]
```

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(2, 1, height_ratios=[2,1])
ax = fig.add_subplot(gs[0])
ax.scatter(exon_covered['Fractiongt0'], exon_covered['Length'].map(np.log10),
           label='Exons', color='red', alpha=0.4)
ax.scatter(intron_covered['Fractiongt0'], intron_covered['Length'].map(np.log10), 
           alpha=1, label='Introns')
ax.set_ylim([2, 6])
ax.set_xlim([0, 1])
ax.set_ylabel('Length of Feature (log10)', fontdict={'size': 14})
ax.set_title('Coverage Consistency Across Feature for %s' % 'Neuronal nucleus-2', fontdict={'size':16})
legend = ax.legend()
for text in legend.get_texts():
    text.set_size(12)

ax.set_xticklabels('')
ax = fig.add_subplot(gs[1])
intron_covered['Fractiongt0'].hist(ax=ax,)
exon_covered['Fractiongt0'].hist(color='red', alpha=0.6, ax=ax)
ax.set_xlabel('Fraction of Feature Covered with >0 Reads', fontdict={'size': 14})
ax.set_ylabel('Number of Features', fontdict={'size':14})
fig.tight_layout()
fig.savefig('../../figures/Coverage_Consistency_Features_Fraction_Covered_Neuronal_Nucleus_2.png')
sns.despine()
```

```python
fig = plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(2, 1, height_ratios=[2,1])
ax = fig.add_subplot(gs[0])
ax.scatter(intron_covered['Reads'].map(log10), intron_covered['Length'].map(np.log10), 
            label='Introns')
ax.scatter(exon_covered['Reads'].map(log10), exon_covered['Length'].map(np.log10),
           label='Exons', color='red', alpha=0.3)

ax.set_ylim([2, 6])
ax.set_xlim([0, 3])
ax.set_ylabel('Length of Feature (log10)', fontdict={'size': 14})
ax.set_title('Coverage Consistency Across Feature for 2B1HCNPN111314', fontdict={'size':16})
legend = ax.legend()
for text in legend.get_texts():
    text.set_size(12)
ax.set_xticklabels('')
ax = fig.add_subplot(gs[1])
intron_covered['Reads'].map(log10).hist(range=(0,3), ax=ax,)
exon_covered['Reads'].map(log10).hist(range=(0,3), color='red', alpha=0.3, ax=ax)
ax.set_xlabel('Number of Reads Mapping to Feature (log10)', fontdict={'size': 14})
ax.set_ylabel('Number of Features', fontdict={'size':14})
fig.tight_layout()
fig.savefig('Coverage_Consistency_Feature_Reads_Mapping_2B1HCNPN111314.png')
```
