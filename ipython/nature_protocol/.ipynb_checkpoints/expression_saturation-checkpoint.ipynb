```python
tpm_df = data.filter_df(data.load_tpms(), genes_only=True)
```

```python
from collections import defaultdict
```

```python
sample_genes = defaultdict(list)
for sample in tpm_df.columns:
    sample_df = tpm_df[tpm_df[sample] > 0][[sample]]
    sample_df = sample_df.applymap(log2)
    for i in range(1, 10):
        in_range = sum((sample_df > i-1) & (sample_df < i)).values[0]
        sample_genes[sample].append(in_range)
    in_range = sum((sample_df > i)).values[0]
    sample_genes[sample].append(in_range)
```

```python
pd.DataFrame(sample_genes).T.plot(kind='barh', stacked=True)
```
