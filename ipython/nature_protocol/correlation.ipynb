```python
tpm_df = data.load_tpms()
```

```python
neuronal_tpm = data.filter_df(tpm_df, relevant_genes=True, non_neuronal=False, pooled=False).applymap(lambda x: log2(x+0.01))
non_neuronal_tpm = data.filter_df(tpm_df, relevant_genes=True, neuronal=False, pooled=False).applymap(lambda x: log2(x+0.01))
pooled_tpm = data.filter_df(tpm_df, relevant_genes=True, neuronal=False, non_neuronal=False).applymap(lambda x: log2(x+0.01))
```

```python
control = '3DHCN100pg111314'
control_name = get_sample_map()[control]
low, mid, high = data.get_low_mid_high_genes(tpm_df[control_name])
control_expressed = tpm_df[tpm_df[control_name] > 0][control_name]
```

```python
def tag_gene(gene):
    if gene in low.index:
        return 'Low'
    if gene in mid.index:
        return 'Mid'
    if gene in high.index:
        return 'High'
    return 'Novel'
```

```python
palette = {'Low': 'Blue',
           'Mid': 'Green',
           'High': 'Red',
           'Novel': 'Purple'}
```

```python
import matplotlib as mpl
```

```python
fig = plt.figure(figsize=(8,4))
low_ax = fig.add_subplot(131)
mid_ax = fig.add_subplot(132)
high_ax = fig.add_subplot(133)
fig.suptitle('Correlation Stratified by Expression in %s' % control_name)
low_ax = sns.corrplot(tpm_df.ix[low.index].sort(axis=1), ax=low_ax, annot=False, diag_names=False, cmap_range=(0, 1), cbar=False)
low_ax.set_title('Low')
mid_ax = sns.corrplot(tpm_df.ix[mid.index].sort(axis=1), ax=mid_ax, annot=False, diag_names=False, cmap_range=(0, 1), cbar=False)
mid_ax.set_title('Mid')
high_ax = sns.corrplot(tpm_df.ix[high.index].sort(axis=1), ax=high_ax, annot=False, diag_names=False, cmap_range=(0, 1), cbar=False)
high_ax.set_title('High')
mid_ax.set_yticklabels([])
high_ax.set_yticklabels([])
```

### Looking at the correlation of TPM across all genes for the 6 neuronal cells.

```python
neuronal_tpm['Expression in %s' % control_name] = neuronal_tpm.index.map(tag_gene)
sns.pairplot(neuronal_tpm,
             hue='Expression in %s' % control_name,  hue_order=['Low','Mid','High','Novel'], 
             palette=palette,
             plot_kws={'alpha': 0.4}, )
neuronal_tpm.groupby('Expression in %s' % control_name).corr(method='spearman')
```

### Let's do it for the other two groups: Non-neuronal and pooled.

```python
non_neuronal_tpm['Expression in Control'] = neuronal_tpm.index.map(tag_gene)
sns.pairplot(non_neuronal_tpm, hue='Expression in Control', hue_order=['Low','Mid','High','Novel'], plot_kws={'alpha': 0.4}
             )
non_neuronal_tpm.groupby('Expression in Control').corr(method='spearman')
```

```python
pooled_tpm['Expression in Control'] = neuronal_tpm.index.map(tag_gene)
sns.pairplot(pooled_tpm, hue='Expression in Control', hue_order=['Low','Mid','High','Novel'], plot_kws={'alpha': 0.4}
             )
pooled_tpm.groupby('Expression in Control').corr(method='spearman')
```
